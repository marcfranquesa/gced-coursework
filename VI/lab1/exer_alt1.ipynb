```python
source = data.seattle_weather()

base = alt.Chart(source).encode(
    x='sum(precipitation):Q',
    y="year:N",
    text="sum(precipitation)"
).transform_calculate(
    year='year(datum.date)'
)

(
    base.mark_bar() +
    base.mark_text(align='left', dx=2)

).properties(title = 'Precipitation per year')
```

```python
source = data.la_riots()


alt.Chart(source).mark_bar().encode(
    x='count():N',
    y=alt.Y('gender:N', axis = alt.Axis(title='', labels=False)),
    color=alt.Color('gender',
                   scale=alt.Scale(
            domain=["Female", "Male"],
            range=['blue', 'green'])),
    row='race:O'
).properties(title = 'Reported cases per gender and race')
```

```python
source = data.unemployment_across_industries()

alt.Chart(source).mark_point(filled=True).transform_aggregate(
     sumCount = 'sum(count):Q',
     groupby=['year', 'series']
).encode(
    x = alt.X('year:O'),
    y = alt.Y('series:N'),
    size=alt.Size('sumCount:Q'),
).properties(title ='Unemployment per year per area')
```

```python
source = data.barley()


base = alt.Chart(source).mark_bar().encode(
    y = alt.Y('year:O'),
    x = alt.X('yield:Q')
)

line = alt.Chart(source).mark_rule(color = 'red').encode(
    x = alt.X('mean(yield):Q', axis = alt.Axis(title = 'yield, mean(red)' ))
)

alt.layer(base, line
          ).facet(column='variety' ,  row = 'site'
          ).transform_filter({'field': 'variety', 'oneOf': ["Glabron", "Manchuria"]}
          )
```
